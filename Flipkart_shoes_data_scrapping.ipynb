# Scrapping Flipkart Shoes Data

```python
import requests
url = 'https://www.flipkart.com/mens-footwear/sports-shoes/pr?sid=osp,cil,1cu&otracker=nmenu_sub_Men_0_Sports%20Shoes'
r = requests.get(url)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.content,'html5lib')
```

```python
#print(soup.prettify)
```

```python
name = soup.find_all('div' ,attrs={'class':'_2WkVRV'},text=True)
```

```python
Name = []
for i in name:
    Name.append(i.text)
```

```python
specifications = soup.find_all('a', attrs={'class':'IRpwTa'},text=True)
```

```python
Spec = []
for i in specifications:
    Spec.append(i.text)
```

```python
price = soup.find_all('div', attrs={'class':'_30jeq3'},text=True)
```

```python
Pr = []
for i in price:
    Pr.append(i.text)
```

```python
discount = soup.find_all('div', attrs={'class':'_3Ay6Sb'},text=True)
```

```python
Disc = []
for i in discount:
    Disc.append(i.text)
```

```python
img = soup.find_all( attrs={'class':'_2r_T1I'})
```

```python
Imag= []
for i in img:
    Imag.append(i.get('src'))
```

```python
import pandas as pd
```

```python
data = {'Name_of_product':Name,'Specifications':Spec,'Price':Pr,'Discount':Disc,'Image_url':Imag}
```

```python
df = pd.DataFrame(data)
```
